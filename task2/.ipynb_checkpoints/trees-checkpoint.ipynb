```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

DATA_PATH = 'train.csv'
```

# Считываем данные

```python
df = pd.read_csv(DATA_PATH)
df.head()
```

Преобразуем дату, разбив на части

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df['hour'] = df.datetime.dt.hour
df['day'] = df.datetime.dt.day
df['month'] = df.datetime.dt.month
df['year'] = df.datetime.dt.year
df.drop(['datetime'], axis=1, inplace=True)
```

```python
df.head()
```

Смотрим матрицу корреляций

```python
import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True)
```

Как видим, с counts сильно коррелируют registered и casual
Так же month коррелирует с season, а atemp - с temp 

```python
df.drop(['registered', 'casual', 'atemp', 'month',],  axis = 1, inplace = True)
df.head()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True)
```

# Создаем и обучаем модель

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, export_graphviz
from sklearn.metrics import mean_squared_error, f1_score
```

Для проверки зарезервируем 1/10 всех данных

```python
y = df['count']
x = df.drop(['count'], axis=1)
x_train, x_test, y_train, y_test = train_test_split(x.values, y, test_size=0.1, random_state=666)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

Для начала выставим максимальную высоту дерева 6

```python
model = DecisionTreeRegressor(max_depth=6, random_state=666)
model.fit(x_train, y_train)
```

Проверим результаты

```python
y_pred = model.predict(x_test)
model.score(x_test, y_test)
```

Среднеквадратичная ошибка

```python
mean_squared_error(y_test, y_pred)
```

```python
export_graphviz(model, out_file='./tree.dot', feature_names=x.columns, filled = True)
# dot -Tpng tree.dot -o tree.png
```

![](tree.png)

# Тюним гиперпараметры

```python
from sklearn.model_selection import GridSearchCV
params = {'max_depth': list(range(3,12)),
          'max_features': list(range(1,11)),
          'criterion' : ['mse', 'mae']}
```

```python
model_grid = GridSearchCV(model, params, cv=5, n_jobs=-1, verbose=True)
model_grid.fit(x_train, y_train)
```

Наилучший результат достигнут при следующих параметрах:

```python
model_grid.best_params_
```

Точность:

```python
model_grid.score(x_test, y_test)
```

Среднеквадратичная ошибка:

```python
y_pred_grid = model_grid.predict(x_test)
mean_squared_error(y_test, y_pred_grid)
```

```python
export_graphviz(model_grid.best_estimator_, out_file='./grid_tree.dot', feature_names=x.columns, filled = True)
# dot -Tpng tree.dot -o tree.png
```
